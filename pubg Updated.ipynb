# Removing Outliers


```python
pubg.boxplot(figsize=(20,10))
plt.xticks(rotation = 45)
```

```python
def remove_outliers_iqr(df, column):
   
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df
numeric_columns = pubg.select_dtypes(include='number').columns
print(numeric_columns)
for i in numeric_columns :
    pubg = remove_outliers_iqr(pubg,i)
```

```python
# Splitting the 'Fire Mode' column, stripping whitespace, and converting to lowercase
pubg['Fire Mode'] = pubg['Fire Mode'].str.strip().str.lower()

# Splitting the values by comma and creating a new DataFrame
split_fire_modes = pubg['Fire Mode'].str.split(',', expand=True)

# Assigning each split part to the corresponding column
pubg['Fire Mode1'] = split_fire_modes[0]
pubg['Fire Mode2'] = split_fire_modes[1]
```

## Fire Modes Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file

# Clean up the 'Fire Mode' column to remove any leading/trailing spaces and convert to lowercase
pubg['Fire Mode'] = pubg['Fire Mode'].str.strip().str.lower()

# Count the occurrences of each fire mode
firemode_counts = pubg['Fire Mode'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
firemode_counts.plot(kind='bar', color='skyblue')
plt.title('Fire Mode Distribution')
plt.xlabel('Fire Mode')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
